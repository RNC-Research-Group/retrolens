```python
import pandas as pd
from glob import glob
import os
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 130)
df = pd.read_csv("shoreline_image_matching.csv")
df
```

```python
retro = df[df.filename.str.startswith("Retrolens")].sample(10)
retro
```

```python
def get_csv(filename):
    bits = filename.split("/")
    terminator = min(bits.index("Stack"), bits.index("Shorelines"))
    bits = bits[:terminator]
    return f"ressci201900060-RNC2-Coastal/{'/'.join(bits)}/{bits[-1]}.csv"

retro["csv_path"] = retro.filename.apply(get_csv)
retro["csv_exists"] = retro.csv_path.apply(os.path.exists)
retro
```
